```python
#2a 
class A:
    def __init__(s, a1, a2):
        s.Aa1 = a1
        s.Aa2 = a2 
    def pA(s):
        print("Aa1: "+str(s.Aa1),"Aa2: "+s.Aa2)
class B(A):
    def __init__(s, b, a1, a2):
        A.__init__(s, a1, a2)
        s.Bb= b
    def pB(s):
        s.pA()
        print("Bb: "+s.Bb)
print("\nInformation displayed:\n")

for i in range(3):
    x=B("Bb"+str(i+1), i, "Aa"+str(i+2))
    x.pB() 
```
